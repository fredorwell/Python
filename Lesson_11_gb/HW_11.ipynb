```python
# f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30
# Определить корни
# Найти интервалы, на которых функция возрастает
# Найти интервалы, на которых функция убывает
# Построить график
# Вычислить вершину
# Определить промежутки, на котором f > 0
# Определить промежутки, на котором f < 0
from numpy import *
import matplotlib.pyplot as plt

#Служебные переменные
limit = 25
step = 0.01
increase_high = True
color = 'r'
line_style = ':'
x_change = {-limit: 'inc'}
x = arange(-limit, limit, step)

def f(x):
    result = -12 * x ** 4 * sin(cos(x)) - 18 * x ** 3 + 5 * x ** 2 + 10 * x - 30
    return result

def switch_color():
    global color
    if color == 'r':
        color = 'b'
    else:
        color = 'r'
    return color

def switch_line():
    global line_style
    if line_style == ':':
        line_style = '-'
    else:
        line_style = ':'
    return line_style
```

```python
plt.plot(x, f(x), 'r', label="-12 * x ** 4 * sin(cos(x)) - 18 * x ** 3 + 5 * x ** 2 + 10 * x - 30")
plt.legend()
plt.grid()
plt.show()
# Поскольку наша функция бесконечна и имеет бесчисленное количество корнней, мы рассмотрим 
# ее на определенном промежутке и на графике отоброзим все изменения в поведении функции
```

```python
x_min = -limit
f_min = f(-limit)
for x_cur in x:
    if f(x_cur) < f_min:
        f_min = round(f(x_cur), 2)
        x_min = round(x_cur, 2)
print(f'Минимальное значение на отрезке от x = {limit} до x = {-limit} - x: {x_min}, y: {f_min}')
```

```python
for i in range(len(x)-1):
    if increase_high:
        if f(x[i]) > f(x[i+1]):
            increase_high = False
            x_change[x[i]] = 'inc'
    else:
        if f(x[i]) < f(x[i+1]):
            increase_high = True
            x_change[x[i]] = 'inc'

x_change[limit] = 'inc'

for i in range(len(x)-1):
    if (f(x[i]) > 0 and f(x[i+1]) < 0) or (f(x[i]) < 0 and f(x[i+1]) > 0):
        x_change[x[i]] = 'zero'
        print(f'На отрезке от x = {limit} до x = {-limit} Корень ф-ции: {x[i]}')
# Поскольку у нас у ф-ции много корней желательно на больших отрезках принт выше коментить
```

```python
x_keys = [x for x in x_change]
x_keys.sort()
# print(x_keys)
for i in range(len(x_keys) - 1):
    x_cur = arange(x_keys[i], x_keys[i+1]+step, step)
    if x_change.get(x_keys[i]) == 'zero':
        switch_line()
    elif x_change.get(x_keys[i]) == 'inc':
        switch_color()
    plt.plot(x_cur, f(x_cur), color=color, linestyle=line_style)
plt.grid()
plt.show()
# Ниже представлен график функции в котором отмечены:
# - Синим цветом периоды возрастания ф-ции
# - Красным периоды убывания ф-ции
# - Сплошной линией отмечены промежуток ф-ции когда она больше нуля
# - Точками отмечены промежутки ф-ции когда она меньше нуля    
```
